# Python Analysis Tasks

Load dataset using Pandas

```python
import pandas as pd

df = pd.read_excel("students.xlsx")

df.head()
```

Create column average_marks

```python
df["average_marks"] = (df["math"] + df["science"] + df["programming"]) / 3

df.head()
```

Identify top-performing student

```python
top_student = df.loc[df["average_marks"].idxmax()]

print("Top Performing Student:")
print(top_student)
```

Group data by department

```python
grouped = df.groupby("department")

print(grouped)
```

Calculate average marks by department

```python
dept_avg = df.groupby("department")["average_marks"].mean()

print("Department-wise Average:")
print(dept_avg)
```

 Analyze relationship between attendance and marks

```python
correlation = df["attendance"].corr(df["average_marks"])

print("Correlation between Attendance and Marks:", correlation)
```

# Visualisation

Bar chart → Student average marks

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df["name"], df["average_marks"])
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Student Average Marks")
plt.xticks(rotation=45)
plt.show()
```

Column Chart → Department Performance

```python
dept_avg = df.groupby("department")["average_marks"].mean()

plt.figure()
plt.bar(dept_avg.index, dept_avg.values)
plt.xlabel("Department")
plt.ylabel("Average Marks")
plt.title("Department Performance")
plt.show()
```

Scatter Plot → Attendance vs Marks

```python
plt.figure()
plt.scatter(df["attendance"], df["average_marks"])
plt.xlabel("Attendance")
plt.ylabel("Average Marks")
plt.title("Attendance vs Marks")
plt.show()
```
